```python
import seaborn as sns
```

```python
#if we have two features called bivariant analysis(verified using jointplot), one featues called univariant analysis, if more than 2 feature use pair plot
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

## Correlation with Heatmap
#### A coorelation heatmap uses coloured cells ,typically in a monochromatic scale,to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection.

```python
# Filter out non-numeric columns from the DataFrame
numeric_df = df.select_dtypes(include='number')

# Compute the correlation matrix for numeric columns
correlation_matrix = numeric_df.corr()

# Display the correlation heatmap
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix)
```

## Jointplot
#### A joint plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatter plot, ahexbin plot, a 2D histogram or 2D density plot.
### Unvariable Analysis

```python
sns.jointplot(x ='tip', y='total_bill', data=df, kind='hex')
```

## Pair plot
#### A pairplot is also known as scatter plot in which one variable in the same data row is matched with another variable's value like this: Pair plot are just elaborations on this, showing at variables paired with all the other variables.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

## Dist plot
#### used to check the distribution of the column feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=True,bins=10)  #kde is kernel density estimation
```

# Categorical Plots
### Seaborn also helps in doing the Categorical Data points. In this section we will discuss about.
#### boxplot
#### violin plot
#### countplot
#### bar plot

```python
##count plot
sns.countplot(y='sex',data=df)  #we can give either one axis on countplot, another axis will be count by default 
```

```python
## barplot
sns.barplot(x='total_bill', y='sex', data=df) #change in the position of x and y will also change the orientation of the bar plot
```

```python
##bar plot
sns.barplot(x='sex', y='total_bill', data=df)
```

## Box plot
#### A box and a whisker plot sometimes called a box plot is a graph that presents info. from a five number summary

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
#categorize my data based on some other categories
sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

## Vioin Plot 
#### Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
