```python
#compute the points on a sin curve
x = np.arange(0, 4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")


#plot the points using matplotlib
plt.plot(x,y)
```

```python
#Subplot
#Compute the x and y coordinates for points on the sine and cosine curve
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#set up a subplot grid that has height 2 and width 1
plt.subplot(2,1,1)

#make the first subplot
plt.plot(x, y_sin)
plt.title("sine")

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cosine')
```

## Bar plot


```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x, y, align='center')
plt.bar(x2, y2, color='g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## Histogram

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('histogram')
```

## Box plot Matplotlib

```python
data=[np.random.normal(0,std, 100) for std in range(1,4)]
#rectangular box plot
#circle here arfe outliers
plt.boxplot(data,  vert=True, patch_artist=True); #vert for orientation and patch_artist for colour
```

## Pie chart

```python
#Data to plot
labels='python','c++','Ruby', 'java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)

plt.axis('equal')
```
